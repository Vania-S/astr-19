```python
#import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
y1= np.sin(x)
y2= np.cos(x)
```

```python
f, graphs = plt.subplots(1,2)          #named the figure graphs but it can be anything
graphs[0].plot(x,y1, color = 'dodgerblue') #plot the domanin and function and change the color
graphs[0].set_xlabel('X')             #label as x
graphs[0].set_ylabel('sin(x)')       # sinx is the function we graphs
graphs[0].set_title('Sin(x)')         # add a title even tho we know what is being graphed

#do the same with graph of cosine
graphs[1].plot(x,y2, color='darkviolet') 
graphs[1].set_xlabel('x')
graphs[1].set_ylabel('cos(x)')
graphs[1].set_title('Cos(x)')

#adjust the distance between the subplots
f.subplots_adjust(wspace= 0.35)
plt.savefig('session_8.pdf')  # save
```
