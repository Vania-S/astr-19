```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.triangular(0,0.2,1,1000)  #triangular (low value, peak value, high value, # of values 
# create the bins
width = 0.01 
histmin = np.floor(min(x))  
histmax = np.ceil(max(x)) + width
    
bins = np.arange(histmin,histmax,width) 
plt.hist(x,bins=bins, alpha = 0.7, edgecolor = 'red')
plt.ylabel("n per bin")
plt.xlabel("x")
```
