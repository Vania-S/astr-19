```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.uniform(low= 0, high =1, size=1000) # change it to uniform so distribution is uniform
# low and high give us range and size is number   # find out from
# create the bins
width = 0.01 # by making the width 0.01, you distribute x bins by 0.01 of 1 which give you 100 bins
histmin = np.floor(min(x))  
histmax = np.ceil(max(x)) + width
    
bins = np.arange(histmin,histmax,width) # needs to this order
plt.hist(x,bins=bins, alpha = 0.65, edgecolor = 'purple')
plt.ylabel("n per bin")
plt.xlabel("x")
```
